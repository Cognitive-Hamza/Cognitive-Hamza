```python
# Welcome
print("Welcome to the Daily Water Intake Tracker!")
print("#_#_#_#_#_#_#_#_#_#_#_#_#_#_#_#_#_#_#_#_#_#")

# Name:
name = input("Please enter your name: ")
if name == "":
	print("name cannot be empty. Please re-run the program.")
else:
	print("Hello,", name)
	print("Let's begin tacking your water intake of the 7 days.\n")

	intake = []
	total_intake = 0

	# loop for 7 days
	for day in range(1, 8):
		glasses = input("Day " + str(day) + ": How much glasses of water did you drink? ")

		if glasses.isdigit():
			glasses = int(glasses)
			intake.append(glasses)
			total_intake = total_intake + glasses
		else:
			print("Invalid input!, it will be counted as 0.")
			intake.append(0)

	average_intake = total_intake / 7
	average_intake = round(average_intake)

	def feedback(avg):
		print("\n|_Final Results_|")
		print("_^_^_^_^_^_^_^_^_^_^_")
		print("Total glasses of water in 7 days:", total_intake)
		print("Average intake of one day:", avg)

		if avg >= 8:
			print("Excellent job! You're staying well hydrated")
		elif avg >= 5:
			print("Good effort! Try to drink a bit more.")
		elif avg == 0:
			print("HOW ARE YOU ALIVE?")
			print("Drink Water!")
		else:
			print("You have to drink more, your body needs it.")

	feedback(average_intake)
	print("Thank you", name, "for using the program!")
```
